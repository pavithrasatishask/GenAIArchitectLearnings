```python
#from langchain.embeddings import OpenAIEmbeddings - this is deprecated now
from langchain_openai import OpenAIEmbeddings
openai_embed = OpenAIEmbeddings(openai_api_key="")
textembed = "hello, welcome to langchain learning"
embeded = openai_embed.embed_query(textembed)
embeded
len(embeded)
print(embeded[:5])
# lenght of the embded vector is 1536 always, this doesnt vary based on the text size. 
#Because embeddings represent meaning, not word-by-word mapping.
#Example:
#"hello" → vector of size 1536
#"hello world" → vector of size 1536
#"hello, welcome to langchain learning" → vector of size 1536
#The length does NOT change based on text size.
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
#from langchain.vectorstores import FAISS - deprecated one
from langchain_community.vectorstores import FAISS

text = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]
huggingface_embed = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = FAISS.from_texts(texts=text, embedding=huggingface_embed)
vectorstore.save_local("faiss_index")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
#from langchain.vectorstores import FAISS - deprecated one
from langchain_community.vectorstores import Chroma

text = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]
huggingface_embed = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = Chroma.from_texts(texts=text, embedding=huggingface_embed,persist_directory="chroma_index")
vectorstore.persist()
```

```python
retriever = vectorstore.as_retriever()
query = "third document?"
#answer = retriever._get_relevant_documents(query)
#The issue is that you're using a private method _get_relevant_documents() which requires additional parameters. You should use the public method instead. Here's the fix:
answer = retriever.invoke(query)
answer
for line in answer:
    print(line.page_content)
```
