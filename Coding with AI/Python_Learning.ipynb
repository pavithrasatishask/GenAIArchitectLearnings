```python
#Data type learning - List
fruit= ["apple", "custard apple", "dragon fruit","banana"]
print(fruit)
fruit
fruit.append("grape")
print(fruit)
fruit.remove("dragon fruit")
print(fruit[2])
print(fruit)
fruit[1] = "papaya" # replace the content in the 1st index
print(fruit)
print(len(fruit))
```

```python
#Data type learning - Tuple
color = ("red","yellow","pink")
print(type(color))
print(color[1])
#color[2] = "blue" # TypeError: 'tuple' object does not support item assignment
```

```python
#Data type learning - Dictionary - KeyValue Pair
studentDetails = {
    "name":"Pavithra",
    "age":34,
    "Native":"Chennai",
    "City":"Chennai"
}
type(studentDetails)
studentDetails["Occupation"] = "IT"
print(studentDetails)
del studentDetails["Occupation"]
print(studentDetails)
print(len(studentDetails))
print(studentDetails["age"])
```

```python
#Data type learning - Set
test = {1,4,6,2}
type(test)
test.add(5)
test.remove(4)
print(test)
#print(test[1]) #TypeError: 'set' object is not subscriptable
```

```python
# Function Learning

def add(a,b):
  total = a+b
  return total

def add_average(a,b):
  avg = a+b/2
  return avg

x=4
y=5
sum = add(x,y)
print(sum)
average = add_average(x,y)
print(average)
```

```python
# Error Handling Concept - try except finally

try:
  #div = 10/0   #ZeroDivisionError: division by zero
  div = 4/2
  print(div)
except Exception as e:
  raise e
finally:
  print("Validations are complete")
```
